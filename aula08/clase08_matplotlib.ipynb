```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt    
```

```python
empresas = ['Apple', 'Google', 'Amazon', 'Microsoft', 'Tencent', 'McDonalds', 'Visa', 'Facebook', 'Alibaba', 'Louis Vuitton']
```

```python
bilhoes = [947.062, 819.573, 705.646, 611.46, 214.023, 196.526, 191.032, 186.421, 169.966, 124.273]
```

```python
plt.plot(empresas, bilhoes)
```

```python
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
```

```python
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
plt.xlabel('Empresas')
plt.ylabel('Bilhões')
```

```python
#Redimensionando
plt.rcParams['figure.figsize'] = (15, 8)

plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
plt.xlabel('Empresas')
plt.ylabel('Bilhões')
```

```python
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
```

```python
#mudando a cor e legendando
plt.plot(empresas, bilhoes, color ='green')
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.legend(['Bilhões de dólares'], loc='best', fontsize=15)
```
